# Import Library

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl
```

# Mendefinisikan variabel input (Antecedent) dan output (Consequent)

```python
quality = ctrl.Antecedent(np.arange(0, 11, 1), 'quality')   # Nilai 0 sampai 10
service = ctrl.Antecedent(np.arange(0, 11, 1), 'service')   # Nilai 0 sampai 10
tip = ctrl.Consequent(np.arange(0, 26, 1), 'tip')           # Nilai 0 sampai 25
```

# Membuat fungsi keanggotaan (Membership Functions)

```python
quality['poor'] = fuzz.trimf(quality.universe, [0, 0, 5])
quality['average'] = fuzz.trimf(quality.universe, [0, 5, 10])
quality['good'] = fuzz.trimf(quality.universe, [5, 10, 10])

service['poor'] = fuzz.trimf(service.universe, [0, 0, 5])
service['average'] = fuzz.trimf(service.universe, [0, 5, 10])
service['good'] = fuzz.trimf(service.universe, [5, 10, 10])

tip['low'] = fuzz.trimf(tip.universe, [0, 0, 13])
tip['medium'] = fuzz.trimf(tip.universe, [0, 13, 25])
tip['high'] = fuzz.trimf(tip.universe, [13, 25, 25])
```

# Menampilkan grafik keanggotaan

```python
quality.view()
service.view()
tip.view()
```

# Mendefinisikan aturan fuzzy (Fuzzy Rules)

```python
rule1 = ctrl.Rule(quality['poor'] | service['poor'], tip['low'])
rule2 = ctrl.Rule(service['average'], tip['medium'])
rule3 = ctrl.Rule(service['good'] | quality['good'], tip['high'])
```

# Membuat kontrol sistem fuzzy

```python
tipping_ctrl = ctrl.ControlSystem([rule1, rule2, rule3])
tipping_simulation = ctrl.ControlSystemSimulation(tipping_ctrl)
```

# Memberikan input dan menjalankan simulasi

```python
tipping_simulation.input['quality'] = 1.0
tipping_simulation.input['service'] = 0.5

# Proses fuzzy (defuzzification)
tipping_simulation.compute()
```

# Menampilkan hasil output dan visualisasinya

```python
print("Output tip yang diberikan: {:.2f}".format(tipping_simulation.output['tip']))
tip.view(sim=tipping_simulation)
```

```python
# Import module yang diperlukan
import skfuzzy as fuzz
import numpy as np
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
# 1. Definisi Variabel Input dan Output
quality = ctrl.Antecedent(np.arange(0, 11, 1), 'quality') # Input 1
(0-10)
tip = ctrl.Antecedent(np.arange(0, 26, 1), 'tip') # Input 2
(0-25)
service = ctrl.Consequent(np.arange(0, 11, 1), 'service') # Output (0-10)
```

```python
# 2. Fungsi Keanggotaan Input (Quality)
quality['poor'] = fuzz.trapmf(quality.universe, [0, 0, 2, 5])
quality['good'] = fuzz.trapmf(quality.universe, [0, 4, 6, 10])
quality['excellent'] = fuzz.trapmf(quality.universe, [5, 8, 10, 10])
quality.view()
```

```python
# 3. Fungsi Keanggotaan Input (Tip)
tip['low'] = fuzz.trimf(tip.universe, [0, 0, 13])
tip['medium'] = fuzz.trimf(tip.universe, [0, 13, 25])
tip['high'] = fuzz.trimf(tip.universe, [13, 25, 25])
tip.view()
```

```python
# 4. Fungsi Keanggotaan Output (Service)
service['poor'] = fuzz.trapmf(service.universe, [0, 0, 3, 5])
service['average'] = fuzz.trapmf(service.universe, [2, 4, 7, 10])
service['excellent'] = fuzz.trapmf(service.universe, [5, 10, 10, 10])
service.view()
```

```python
# 5. Aturan Fuzzy Rules
rules = [
 ctrl.Rule(quality['poor'] & tip['low'], service['poor']),
 ctrl.Rule(quality['poor'] & tip['medium'], service['poor']),
 ctrl.Rule(quality['poor'] & tip['high'], service['average']),

 ctrl.Rule(quality['good'] & tip['low'], service['average']),
 ctrl.Rule(quality['good'] & tip['medium'], service['average']),
 ctrl.Rule(quality['good'] & tip['high'], service['excellent']),

 ctrl.Rule(quality['excellent'] & tip['low'], service['average']),
 ctrl.Rule(quality['excellent'] & tip['medium'],
service['excellent']),
 ctrl.Rule(quality['excellent'] & tip['high'], service['excellent'])
]
```

```python
# 6. Sistem Kontrol
service_ctrl = ctrl.ControlSystem(rules)
service_sim = ctrl.ControlSystemSimulation(service_ctrl)
```

```python
# 7. Input dan Simulasi
service_sim.input['quality'] = 7.0 # Contoh input quality (0-10)
service_sim.input['tip'] = 18.0 # Contoh input tip (0-25)
service_sim.compute() # Proses perhitungan
# 8. Hasil dan Visualisasi
print("Predicted Service Level:", service_sim.output['service'])
service.view(sim=service_sim)
```
