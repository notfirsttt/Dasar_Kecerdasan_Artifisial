## Guided 1 (Penggunaan Figure dalam Matplotlib)

```python
    import matplotlib.pyplot as plt

    # Inisialisasi Figure dengan ukuran 8x6
    fig = plt.figure(figsize=(8,6))

    # Membuat subplot dalam figure
    ax = fig.add_subplot(1,1,1)  # 1 baris, 1 kolom, indeks 1

    # Membuat plot sederhana
    ax.plot([1,2,3,4], [10,20,25,30], label='Data1')

    # Menambah label dan judul
    ax.set_xlabel("X Axis")
    ax.set_ylabel("Y Axis")
    ax.set_title("Contoh Plot dalam Figure")

    # Menampilkan legenda
    ax.legend()

    # Menampilkan plot
    plt.show()
```

Kode di atas menggunakan Matplotlib untuk membuat plot sederhana dengan satu Figure dan satu subplot. Pertama, Figure diinisialisasi dengan ukuran delapan kali enam inci lalu ditambahkan subplot menggunakan add subplot satu satu satu. ## Guided 2 (Labeling dan Legenda)

## Guided 3 (Memberikan Anotasi Pada Label)

## Guided 7 (Fill Between)

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4]
y1 = [9, 20, 27, 35]
y_main = [7, 18, 22, 29]  # Data garis utama
y2 = [6, 15, 18, 23]

# Membuat figure
plt.figure(figsize=(10, 6))

# Membuat fill_between plot
plt.fill_between(x, y1, y2, color='pink', alpha=0.5, label='Fill Plot')

# Membuat plot garis utama
plt.plot(x, y_main, color='red', linewidth=2, label='Garis Utama')

# Labeling plot
plt.title("Fill Between")
plt.xlabel("X")
plt.ylabel("Y")

# Menampilkan legend dan figure
plt.legend(loc='lower right') # legenda di kanan bawah
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    "Nama": ["Budi", "Siti", "Asep", "Dewi", "Andi", "Lestari", "Joko", "Rina", "Fajar", "Indah",
             "Agus", "Fitri", "Rudi", "Maya", "Hadi", "Sri", "Wawan", "Yuni", "Bambang", "Ayu"],
    "Algoritma Pemrograman": [81.24, 98.52, 91.96, 87.96, 74.68, 74.68, 71.74, 95.99, 88.03, 91.24,
                              70.62, 99.1, 94.97, 76.37, 75.45, 75.5, 79.13, 85.74, 82.96, 78.74],
    "Matematika Diskrit": [73.66, 84.86, 71.03, 97.28, 77.76, 89.88, 79.35, 85.6, 86.4, 75.55,
                           99.09, 93.25, 98.18, 96.84, 87.94, 97.66, 72.65, 75.88, None, 79.76],
    "Dasar Kecerdasan Artifisial": [95.89, 88.7, 79.93, 71.91, 79.33, 79.76, 91.89, 89.13, 96.62, 84.17,
                                    73.59, 91.4, 92.82, 86.84, 93.13, 84.81, 85.68, 82.83, 70.76, 73.24],
    "Tahun Angkatan": [2023, 2024, 2022, 2024, 2023, 2022, 2022, 2022, 2024, 2023,
                       2022, 2022, 2022, 2024, 2024, 2023, 2024, 2022, 2023, 2022]
}

# Konversi ke DataFrame
df = pd.DataFrame(data)

# Konversi nilai ke numerik (hindari error karena None atau string)
df.iloc[:, 1:-1] = df.iloc[:, 1:-1].apply(pd.to_numeric, errors='coerce')

# Grouping berdasarkan Tahun Angkatan dan agregasi hanya untuk kolom numerik
grouped = df.groupby("Tahun Angkatan")[df.columns[1:-1]].agg(["max", "min", "mean"]) # Only include numeric columns for aggregation

# Plotting
fig, ax = plt.subplots(figsize=(10,6))

# Loop untuk setiap mata kuliah
for subject in df.columns[1:-1]:  # Hanya kolom numerik
    ax.plot(grouped.index, grouped[subject]["max"], marker="o", linestyle="-", label=f"Max {subject}")
    ax.plot(grouped.index, grouped[subject]["min"], marker="s", linestyle="--", label=f"Min {subject}")
    ax.plot(grouped.index, grouped[subject]["mean"], marker="d", linestyle="-.", label=f"Mean {subject}")

# Labeling
ax.set_title("Nilai Maksimum, Minimum, dan Rata-rata per Angkatan")
ax.set_xlabel("Tahun Angkatan")
ax.set_ylabel("Nilai")
ax.legend()
plt.grid()
plt.show()
```

Kode ini bertujuan untuk menganalisis dan memvisualisasikan distribusi nilai mahasiswa dari berbagai angkatan dalam tiga mata kuliah, yaitu **Algoritma Pemrograman, Matematika Diskrit, dan Dasar Kecerdasan Artifisial**. Data awal disimpan dalam bentuk dictionary dan dikonversi menjadi **DataFrame pandas** untuk memudahkan pengolahan. Setelah itu, nilai-nilai dari setiap mata kuliah dikonversi ke format numerik untuk memastikan tidak ada error dalam perhitungan. Selanjutnya, data dikelompokkan berdasarkan **tahun angkatan** dengan melakukan agregasi nilai **maksimum, minimum, dan rata-rata** untuk setiap mata kuliah. Hasil analisis ini kemudian divisualisasikan dalam bentuk **grafik garis**, di mana setiap mata kuliah memiliki tiga garis yang menunjukkan tren nilai maksimum, minimum, dan rata-rata berdasarkan tahun angkatan mahasiswa. Dengan grafik ini, kita dapat dengan mudah mengamati bagaimana performa akademik mahasiswa dari setiap angkatan dalam mata kuliah yang dianalisis, apakah terdapat pola tertentu atau perbedaan signifikan antar angkatan.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Membuat graf berbobot
G = nx.Graph()

# Posisi simpul (kota) yang sudah disesuaikan
posisi = {
    "Oradea": (3, 8), "Zerind": (2.5, 7), "Arad": (1.8, 6), "Timisoara": (2, 3.5),
    "Lugoj": (4, 2.5), "Mehadia": (4.1, 1.25), "Drobeta": (3.5, 0), "Craiova": (6, -0.3),
    "Rimnicu Vilcea": (5.5, 3.3), "Sibiu": (4.8, 5.3), "Fagaras": (7, 4.8), "Pitesti": (7.3, 2.6),
    "Bucharest": (9, 0.7), "Giurgiu": (8.5, -0.8), "Urziceni": (10.3, 1.3), "Hirsova": (12, 1.5),
    "Eforie": (12.4, -0.2), "Vaslui": (11.5, 4.7), "Iasi": (10.4, 6.5), "Neamt": (9.2, 7.5)
}

# Menambahkan edge (jalur antar kota) dan bobotnya
edges = [
    ("Oradea", "Zerind", 71), ("Oradea", "Sibiu", 151), ("Zerind", "Arad", 75),
    ("Arad", "Timisoara", 118), ("Arad", "Sibiu", 140), ("Timisoara", "Lugoj", 111),
    ("Lugoj", "Mehadia", 70), ("Mehadia", "Drobeta", 75), ("Drobeta", "Craiova", 120),
    ("Craiova", "Pitesti", 138), ("Craiova", "Rimnicu Vilcea", 146),
    ("Rimnicu Vilcea", "Sibiu", 80), ("Rimnicu Vilcea", "Pitesti", 97),
    ("Sibiu", "Fagaras", 99), ("Fagaras", "Bucharest", 211), ("Pitesti", "Bucharest", 101),
    ("Bucharest", "Giurgiu", 90), ("Bucharest", "Urziceni", 85), ("Urziceni", "Hirsova", 98),
    ("Hirsova", "Eforie", 86), ("Urziceni", "Vaslui", 142), ("Vaslui", "Iasi", 92),
    ("Iasi", "Neamt", 87)
]

# Menambahkan edge ke dalam graf
G.add_weighted_edges_from(edges)

# Gambar graf dengan node berbentuk kotak kecil kosong
plt.figure(figsize=(10, 6))

# Gambar edge (garis antar simpul)
nx.draw_networkx_edges(G, posisi, edge_color="black", width=1)

# Menampilkan bobot (jarak) di atas garis, bukan di dalam garis
edge_labels = {(u, v): d for u, v, d in edges}
nx.draw_networkx_edge_labels(G, posisi, edge_labels=edge_labels, font_size=8, label_pos=0.6)

# Menggambar node sebagai kotak bold
for kota, (x, y) in posisi.items():
    plt.scatter(x, y, s=130, facecolors="white", edgecolors="black", marker="s", linewidths=2.5)  # Kotak lebih tebal

    # Geser beberapa label kota sesuai permintaan
    offset_x, offset_y = 0.2, 0  # Offset default ke kanan
    if kota in ["Drobeta", "Zerind", "Pitesti"]:
        offset_x = -1.2
    if kota in ["Arad"]:
        offset_x = -0.8
    elif kota in ["Sibiu", "Rimnicu Vilcea", "Timisoara"]:
        offset_y = 0.2
    elif kota in ["Bucharest", "Urziceni", "Pitesti"]:
        offset_y = -0.3  # Geser ke bawah

    plt.text(x + offset_x, y + offset_y, kota, fontsize=9, fontweight="bold", verticalalignment="center")  # Nama bold

# Atur tampilan plot
plt.axis("off")
plt.title("Graf Kota-Kota Besar di Eropa", fontsize=12, fontweight="bold")

# Tampilkan graf
plt.show()
```

Kode ini membuat visualisasi graf berbobot yang merepresentasikan jaringan jalan antar kota besar di Eropa, khususnya di wilayah Rumania. **NetworkX** digunakan untuk membangun struktur graf, sedangkan **Matplotlib** dipakai untuk menampilkan visualisasinya. Setiap kota direpresentasikan sebagai simpul (**node**) dengan koordinat yang telah disesuaikan agar terlihat proporsional. Hubungan antar kota (jalur jalan) direpresentasikan sebagai **edge** dengan bobot yang menunjukkan jarak antar kota dalam satuan tertentu.

Edge atau jalur antar kota digambar menggunakan garis hitam, sementara bobotnya (jaraknya) ditampilkan di atas garis dengan ukuran font kecil. Untuk memperjelas tampilan, simpul atau kota digambar sebagai **kotak tebal berwarna putih dengan tepi hitam**, bukan lingkaran seperti biasanya. Nama kota kemudian ditempatkan di sekitar kotak dengan penyesuaian posisi agar tidak bertumpukan. Akhirnya, tampilan graf dibuat tanpa sumbu (**axis off**) dan diberikan judul yang lebih informatif. Dengan visualisasi ini, kita dapat dengan mudah memahami bagaimana kota-kota dihubungkan dan memperkirakan rute perjalanan berdasarkan bobot yang diberikan.
